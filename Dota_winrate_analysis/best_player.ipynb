```python
for p in response.json()["players"]:
    if not p["steamAccount"]["isAnonymous"]:
        if "proSteamAccount" in p["steamAccount"]:
            name = p["steamAccount"]["proSteamAccount"]["name"]
        else:
            name = ''
        id = p["steamAccount"]["id"]
        imp = p["imp"]
        
        #print(p["imp"], p["steamAccount"]["id"], p["steamAccount"]["proSteamAccount"]["name"])
        req = requests.get(api + str(p["steamAccount"]["id"])).json()
        if "matchCount" in req:
            g = req["matchCount"]
        else:
            g = 0
        win = req["winCount"]
        #print(req["matchCount"], req["winCount"])
        players = players.append({'name' : name,
                                  'id' : id,
                                  'imp' : imp,
                                  'g' : g,
                                  'win' : win},
                                 ignore_index=True)
```
